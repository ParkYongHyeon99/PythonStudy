```python
import pandas as pd
```

```python
import bs4
```

```python
url = 'https://ddragon.leagueoflegends.com/cdn/13.24.1/data/ko_KR/champion.json'
```

```python
a = requests.get(url)
```

```python
res = a.json()
```

# champion_id, champion_name_kr, champion_name_en

```python
list(res['data'].keys())
```

```python
data = res['data']
```

```python
champ_list = list(data.keys())
```

```python
result = list(map(lambda x:[data[x]['key'], data[x]['name'], x] ,champ_list))
```

```python
df = pd.DataFrame(result, columns = ['champion_id', 'champion_name_kr', 'champion_name_en'])
```
