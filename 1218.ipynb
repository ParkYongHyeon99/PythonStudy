```python
def get_puuid(nickname):
    nameS = nickname.split('#')
    url = f'https://asia.api.riotgames.com/riot/account/v1/accounts/by-riot-id/{nameS[0]}/{nameS[1]}?api_key={riot_api_key}'
    res = requests.get(url).json()
    puuid = res['puuid']
    return puuid
```
