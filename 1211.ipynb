주민등록번호 13자리를 문자열로 입력 받아, 받은 주민번호를 기반으로 남성인지 여성인지를 판별하여 반환하는 함수를 작성하시오.

입력 문자열은 "주민번호 앞자리 - 뒷자리"로 구성됨 반환 문자열은 "남성", "여성" 둘중 하나

주민번호 뒷자리중 첫번째 숫자가 1이면 남성, 2이면 여성

예1)

입력(문자열) : "820327-1022421" 반환(문자열) : "남성" 

예2)

입력(문자열) : "820327-2022421" 반환(문자열) : "여성"

```python
def q (a):
    if int(a[7]) == 1:
        return '남자'
    elif int(a[7]) == 0:
        return '여자'
    else:
        return '잘못된 양식입니다.'
```

```python
q('820327-1022421')
```

# kda
## (kill + assists)/deaths
### kda컬럼추가
### none -> apply 함수를 사용해서 kda계산 소수점 둘째자리까지 사용
### kda가 높은게 위로 가도록 sort

```python
df['kda'] = 'none'
```

```python
def kda(x, y, z):
    if y == 0:
        return x+z
    else:
        result = round(((x + z) / y), 2)
        return result
```

```python
def kda(x):
    try:
        kda = round((x.kills + x.assists)/x.deaths,2)
    except:
        kda = (x.kills + x.assists)
    return kda
```

```python
df['kda'] = df.apply(lambda x: kda(x.kills, x.deaths, x.assists), axis = 1)
```

```python
df
```

```python
df.sort_values(by = ['kda'], ascending = False, inplace = True)
```

# dm_per 컬럼

```python
my_df['dm_per'] = 'none'
```

```python
my_df
```

```python
def dm_per(x):
    if x.damageDealt == 0:
        result = x.damageTaken
    elif x.damageTaken == 0:
        result = x.damageDealt
    else:
        result = (x.damageDealt/x.damageTaken)
    return result
```

```python
my_df['dm_per'] = my_df.apply(lambda x : dm_per(x), axis = 1)
```
